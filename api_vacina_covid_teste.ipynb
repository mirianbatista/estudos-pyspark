```python
import requests
from requests.auth import HTTPBasicAuth     
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
r = requests.get('https://imunizacao-es.saude.gov.br/desc-imunizacao/_search', auth=('imunizacao_public', 'qlto5t&7r_@+#Tlstigi'))  
r.json()
```

```python
a = json.loads(r.text)
res = json_normalize(a)
df = pd.DataFrame(res)
df
```

```python
df.rename(columns={"hits.hits": 'hits'}, inplace=True)
```

```python
df
```

```python
q = df[['hits']]
q
```

```python
qq = pd.DataFrame(q['hits'].explode().tolist())
qq
```

```python
a = json_normalize(qq._source)
a.info()
```
